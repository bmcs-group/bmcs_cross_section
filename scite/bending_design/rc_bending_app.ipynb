## 💡 Tips for Exploration

### Understanding the Parameters

**Loading (Orange Section):**
- **M_Ed** - Design bending moment acting on the cross-section
- **N_Ed** - Design axial force (negative = compression, positive = tension)
- The section is subjected to combined axial force and bending moment action

**Geometry (Blue Section):**
- **d** - Effective depth: distance from top fiber to reinforcement centroid
- **b** - Width of the rectangular cross-section
- **c_nom** - Nominal concrete cover protecting the reinforcement

**Materials (Blue Section):**
- **Characteristic Values:** f_ck (concrete strength), f_yk (steel strength)
- **Design Values:** f_cd, f_yd calculated using safety factors γ_c, γ_s
- Observe how characteristic values are converted to design values

**Stress-Strain Profiles (Green Section):**
- Visual representation of strain and stress distributions across the section
- Shows neutral axis position and compression/tension zones
- **ε_s,bottom slider** - Adjust the tension strain at the reinforcement level to explore different strain states

**Assessment Result (Green Section):**
- **M_Rd** - Moment capacity calculated from the current strain state
- **M_Eds** - Moment to be resisted considering second-order effects (M_Ed + N_Ed·z_s1)
- **ΔM** - The major indicator of design feasibility: ΔM = M_Rd - M_Eds
  - ΔM ≥ 0 (green) → Design is safe and economical
  - ΔM < 0 (red) → Design is unsafe, more reinforcement needed
- **State Variables** - Neutral axis (x), lever arm (z), compression force (F_cd), tension force (F_sd)

### Exploring Strain States

1. Start with the default values and observe the stress-strain profiles
2. Use the **ε_s,bottom slider** (below the visualization) to adjust the reinforcement strain
3. Watch how the stress-strain profiles and neutral axis position update in real-time
4. Observe the state variables (x, z, F_cd, F_sd) change as you adjust the strain
5. **Key indicator:** Monitor ΔM in the Assessment Result box - this shows whether your design is feasible, economical, or unsafe
6. Check how the required reinforcement area (A_s) varies with different strain states
   - A_s is calculated from force equilibrium: **A_s = F_sd / f_yd**

### Design Workflow

The tool follows the structural design workflow:
1. **Define loading** → set design actions (M_Ed, N_Ed) for combined loading
2. **Define geometry** → cross-section dimensions
3. **Select materials** → characteristic strengths with safety factors applied
4. **Explore strain states** → use slider to investigate different scenarios
5. **Assess results** → determine required reinforcement and verify capacity through ΔM

### Key Concepts

- **ε_c,top is fixed at -3.5‰** - This represents the ultimate concrete compression strain
- **ε_s,bottom varies** - Explore different reinforcement strain states from elastic to plastic range
- **Neutral axis position (x)** - Automatically calculated from the assumed strain state
- **Reinforcement area (A_s)** - Calculated from force equilibrium: **A_s = F_sd / f_yd**
- **ΔM is the design criterion** - Adjust ε_s,bottom to achieve ΔM ≥ 0 for a safe and economical design
