# Using Component Catalogs

This notebook demonstrates the features of the `cs_components` subpackage:

- **Overview**: Getting catalog overviews
- **Accessing**: Loading and querying component catalogs
- **Parameters**: Retrieving component properties

The `cs_components` package provides a standardized product catalog layer that separates physical products from mathematical material models (matmod).

## 11. Comparing Multiple Steel Grades

Compare stress-strain behavior of different steel grades (B500A, B500B, B500C) for the same diameter.

```python
# Create components for each grade
steel_b500a = SteelRebarComponent(nominal_diameter=16, grade='B500A')
steel_b500b = SteelRebarComponent(nominal_diameter=16, grade='B500B')
steel_b500c = SteelRebarComponent(nominal_diameter=16, grade='B500C')

# Use built-in plotting on shared axes
fig, ax = plt.subplots(figsize=(12, 7))

steel_b500a.plot_stress_strain(ax=ax, show_limits=False, color='blue', 
                                label=f'B500A (ε_ud={steel_b500a.eps_ud:.4f})')
steel_b500b.plot_stress_strain(ax=ax, show_limits=False, color='red',
                                label=f'B500B (ε_ud={steel_b500b.eps_ud:.4f})')
steel_b500c.plot_stress_strain(ax=ax, show_limits=False, color='green',
                                label=f'B500C (ε_ud={steel_b500c.eps_ud:.4f})')

# Add ultimate strain markers manually
ax.axvline(x=steel_b500a.eps_ud * 1000, color='blue', linestyle='--', alpha=0.5)
ax.axvline(x=steel_b500b.eps_ud * 1000, color='red', linestyle='--', alpha=0.5)
ax.axvline(x=steel_b500c.eps_ud * 1000, color='green', linestyle='--', alpha=0.5)

ax.set_xlim(0, 30)
ax.set_title('Comparison of Steel Grades (D16, Design Curves)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\nDuctility comparison (ε_ud):")
print(f"  B500A: {steel_b500a.eps_ud:.4f} (most ductile)")
print(f"  B500B: {steel_b500b.eps_ud:.4f}")
print(f"  B500C: {steel_b500c.eps_ud:.4f} (highest ductility)")
```

## 14. Using Components in Cross-Section Design

Demonstrate how to use catalog components in actual cross-section assembly.

```python
from bmcs_cross_section.cs_design import (
    RectangularShape,
    ReinforcementLayer,
    ReinforcementLayout,
    CrossSection
)

# Select components from catalog
concrete = get_concrete_by_class("C30/37")
steel_rebar = SteelRebarComponent(nominal_diameter=20, grade='B500B')

# Define cross-section geometry
shape = RectangularShape(b=300, h=500)

# Define reinforcement layers using catalog components
# Bottom layer: 4 bars of D20
n_bars_bottom = 4
As_bottom = steel_rebar.area * n_bars_bottom
bottom_layer = ReinforcementLayer(
    z=450,  # mm from top
    A_s=As_bottom,
    material=steel_rebar.matmod
)

# Top layer: 2 bars of D20
n_bars_top = 2
As_top = steel_rebar.area * n_bars_top
top_layer = ReinforcementLayer(
    z=50,  # mm from top
    A_s=As_top,
    material=steel_rebar.matmod
)

# Create reinforcement layout
reinforcement = ReinforcementLayout(layers=[bottom_layer, top_layer])

# Create cross-section
cross_section = CrossSection(
    shape=shape,
    concrete=concrete.matmod,
    reinforcement=reinforcement
)

# Display configuration
print("Cross-Section Configuration:")
print("=" * 60)
print(f"\nConcrete: {concrete.name}")
print(f"  f_ck = {concrete.f_ck} MPa")
print(f"  f_cd = {concrete.f_cd:.2f} MPa")
print(f"  E_cm = {concrete.E_cm} MPa")

print(f"\nReinforcement: {steel_rebar.name}")
print(f"  Diameter: {steel_rebar.nominal_diameter} mm")
print(f"  Area per bar: {steel_rebar.area:.2f} mm²")
print(f"  f_tk = {steel_rebar.f_tk} MPa")
print(f"  f_td = {steel_rebar.f_td:.1f} MPa")

print(f"\nGeometry: Rectangular")
print(f"  Width: {shape.b} mm")
print(f"  Height: {shape.h} mm")

print(f"\nReinforcement Layers:")
print(f"  Bottom: {n_bars_bottom} × D{steel_rebar.nominal_diameter} = {As_bottom:.0f} mm² @ z={bottom_layer.z} mm")
print(f"  Top:    {n_bars_top} × D{steel_rebar.nominal_diameter} = {As_top:.0f} mm² @ z={top_layer.z} mm")

print("\n" + "=" * 60)
print("✓ Cross-section created successfully using catalog components")
print("=" * 60)
```

```python
# Visualize the cross-section
cross_section.plot_cross_section()
plt.title('Cross-Section with Catalog Components', fontsize=13, fontweight='bold')
plt.show()
```

## 15. Summary and Benefits

The component catalog system provides:

### Key Features:
- **Standardization**: Pre-defined products with validated properties
- **Traceability**: Product IDs, manufacturer info, certifications
- **Safety**: Built-in characteristic → design value conversion (γ_s, γ_c)
- **Query Interface**: Pandas-based filtering and searching
- **Integration**: Seamless connection to material models (matmod)
- **Flexibility**: Easy to extend with new products

### Workflow:
1. **Browse catalogs** using `create_XXX_catalog()` functions
2. **Query products** using search functions or DataFrame filtering
3. **Create components** by instantiating component classes
4. **Access parameters** through properties (f_td, eps_ud, etc.)
5. **Visualize behavior** using `get_design_stress_strain()`
6. **Use in design** by linking matmod to ReinforcementLayer
